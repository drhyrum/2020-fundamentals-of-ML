```python
%matplotlib notebook
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
```

```python
data = pd.read_csv('../../data/20questions/micro.csv')
```

```python
display(data)
```

```python
target = data[data.columns[0]]
feature_names = data.columns[1:]
features = data[feature_names]
```

```python
model = DecisionTreeClassifier().fit(features, target)
```

```python
from IPython.display import display
plot_tree(model, feature_names=feature_names, class_names=model.classes_, impurity=False, node_ids=False, label='none', filled=True, fontsize=8)
```
